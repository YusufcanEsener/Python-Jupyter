```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
cat_cols=list(kalp.select_dtypes("object").columns.values)
kalp[cat_cols]=kalp[cat_cols].apply(LabelEncoder().fit_transform)
kalp.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig=plt.figure(figsize=(20,20))
sns.heatmap(kalp.corr(), annot=True)
```

```python
nitelikler=['Age', 'Sex', 'ChestPainType', 'RestingBP', 'Cholesterol', 'FastingBS', 'RestingECG', 'MaxHR', 'ExerciseAngina', 'Oldpeak', 'ST_Slope']

print(len(nitelikler)) #niteliklerin uzunluğu
```

```python
X=kalp[nitelikler] #x'e nitelikleri ata
y=kalp['HeartDisease'] #y'ye hedef sınıfı ata
```

```python
X=kalp.iloc[:,:11]
y=kalp.iloc[:,-1] 
```

```python
from sklearn.model_selection import train_test_split
import sklearn.metrics as accuracy_score 
import sklearn.metrics as metrics

import matplotlib.pyplot as plt

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=215)
dtree = DecisionTreeClassifier(criterion='gini',max_depth=3)

dtree.fit(X_train, y_train)

print('Egitim seti:',dtree.score(X_train,y_train))
print('Test seti:',dtree.score(X_test,y_test))

y_pred = dtree.predict(X_test)

print("Accuracy:",metrics.accuracy_score(y_test, y_pred)*100)

print("\nFeatures importance and coefficient scores")
print(dtree.feature_importances_)
importance = np.abs(dtree.feature_importances_)
feature_names = np.array(dtree.feature_names_in_)
plt.bar(height=importance, x=feature_names)
plt.title("Feature importances via coefficients")
plt.xticks(rotation=30, ha='right')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
import sklearn.metrics as accuracy_score 
import sklearn.metrics as metrics

x = range(1,50)
b = range(1,600)

listedogruluk=[]
listerandomstate=[]
listeneighbors=[]
for n in x:
    for a in b:
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=a)
        dtree =DecisionTreeClassifier(criterion='gini',max_depth=n)
        dtree = dtree.fit(X_train,y_train)
        y_pred = dtree.predict(X_test)
        listerandomstate.append(a)
        listeneighbors.append(n)
        listedogruluk.append(metrics.accuracy_score(y_test, y_pred)*100)
print(max(listedogruluk))
#print(listedogruluk.index(max(listedogruluk)))
print("random state: ",listerandomstate[listedogruluk.index(max(listedogruluk))])
print("n_neighbors: ",listeneighbors[listedogruluk.index(max(listedogruluk))])
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred))

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
import seaborn as sns
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,cmap=plt.cm.YlOrRd)
plt.show()
```

```python
from math import sqrt
from sklearn import metrics
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score

print("Mean Absolute Error: ", metrics.mean_absolute_error(y_test, y_pred))
print("RMSE:",sqrt(metrics.mean_squared_error(y_test, y_pred)))
print("Matthews_CorrCoef (MCC):", metrics.matthews_corrcoef(y_test, y_pred))
print("Cohen_Kappa_Score:", metrics.cohen_kappa_score(y_test, y_pred))
print("AUROC Score:", metrics.roc_auc_score(y_test, y_pred))
print("AUPRC Score:", metrics.average_precision_score(y_test, y_pred))
print("F-Measure:", metrics.f1_score(y_test,y_pred))
print("Precision Score:", precision_score(y_test, y_pred))
print ("Recall Score", recall_score(y_test, y_pred)) #sensitivity Value - tp / (tp + fn)
#sensitivity (true positive rate) and specificity (true negative rate)
#specificity
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
print ("Specificity:", tn / (tn+fp))
print("Accuracy:", metrics.accuracy_score(y_test, y_pred)*100)
```
